````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
````

````python
df = pd.read_csv('amazon.csv')
````

````python
df.head()
````

DATA CLEANING


````python
df.info()
````

````python
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Ship Date'] = pd.to_datetime(df['Ship Date'])
````

````python
df
````

````python
df['order_weekday_name'] = df['Order Date'].dt.strftime('%A')
df['order_month_name'] = df['Order Date'].dt.strftime('%B')
df['order_day_number'] = df['Order Date'].dt.day
df['order_month_number'] = df['Order Date'].dt.month
df['order_year'] = df['Order Date'].dt.year
df['ship_weekday_name'] = df['Ship Date'].dt.strftime('%A')
df['ship_month_name'] = df['Ship Date'].dt.strftime('%B')
df['ship_day_number'] = df['Ship Date'].dt.day
df['ship_month_number'] = df['Ship Date'].dt.month
df['ship_year'] = df['Ship Date'].dt.year
````

````python
df
````

````python
df.drop(columns=['Order Date', 'Ship Date'], inplace=True)
````

````python
df['Order Priority'].unique()
````

````python
mapping = { 'H' : 'High'
           ,'M' : 'Medium'
           ,'L' : 'Low'
           ,'C' : 'Critical' }

df['Order Priority'] = df['Order Priority'].map(mapping)


````

Gross Profit Margin


````python
df['Profit'] = df['Unit Price'] - df['Unit Cost']
profit_by_item = df.groupby('Item Type')['Profit'].first().reset_index()
print(profit_by_item)
````

Profit Margin (%)

````python
df_grouped = df.groupby('Item Type').agg(
    Total_Profit=('Total Profit', 'sum'),
    Total_Revenue=('Total Revenue', 'sum')).reset_index()
df_grouped['Profit_Margin'] = (df_grouped['Total_Profit'] / df_grouped['Total_Revenue']) * 100
df_grouped[['Item Type','Profit_Margin']]
````

Most sold item by each country?

````python

df_grouped = df.groupby(['Country', 'Item Type'], as_index=False)['Units Sold'].sum()
df_sorted = df_grouped.sort_values(by=['Country', 'Units Sold'], ascending=[True, False])
most_sold_items = df_sorted.drop_duplicates(subset='Country', keep='first').reset_index(drop=True)

print(most_sold_items)
````

Least sold item by each country

````python
df_grouped1 = df.groupby(['Country', 'Item Type'], as_index=False)['Units Sold'].sum()
df_sorted1 = df_grouped1.sort_values(by=['Country', 'Units Sold'], ascending=[True, True])
least_sold_items = df_sorted1.drop_duplicates(subset='Country', keep='first').reset_index(drop=True)

print(least_sold_items)
````
